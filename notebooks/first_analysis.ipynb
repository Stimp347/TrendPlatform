```python
# Ячейка 1: Импорт библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from IPython.display import display, HTML

# Настройка отображения
plt.style.use('seaborn-v0_8-darkgrid')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 200)

print("✓ Библиотеки загружены")

# Ячейка 2: Загрузка данных
import os
import glob

# Находим все CSV файлы в data
csv_files = glob.glob("../data/*.csv")
print(f"Найдено файлов: {len(csv_files)}")

# Загружаем все данные
data_frames = {}
for file in csv_files:
    topic = os.path.basename(file).replace('vk_', '').replace('.csv', '')
    df = pd.read_csv(file, encoding='utf-8-sig')
    data_frames[topic] = df
    print(f"  {topic}: {len(df)} записей")

# Ячейка 3: Быстрый анализ
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Анализ собранных данных', fontsize=16)

# 1. Количество постов по темам
post_counts = pd.Series({k: len(v) for k, v in data_frames.items()})
axes[0,0].bar(post_counts.index, post_counts.values, color='skyblue')
axes[0,0].set_title('Количество постов по темам')
axes[0,0].tick_params(axis='x', rotation=45)

# 2. Средние лайки
avg_likes = pd.Series({k: v['likes'].mean() for k, v in data_frames.items()})
axes[0,1].bar(avg_likes.index, avg_likes.values, color='lightcoral')
axes[0,1].set_title('Среднее количество лайков')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Средние репосты
avg_reposts = pd.Series({k: v['reposts'].mean() for k, v in data_frames.items()})
axes[1,0].bar(avg_reposts.index, avg_reposts.values, color='lightgreen')
axes[1,0].set_title('Среднее количество репостов')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Средние комментарии
avg_comments = pd.Series({k: v['comments'].mean() for k, v in data_frames.items()})
axes[1,1].bar(avg_comments.index, avg_comments.values, color='gold')
axes[1,1].set_title('Среднее количество комментариев')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Ячейка 4: Интерактивный график Plotly
if 'кофе' in data_frames and 'чай' in data_frames:
    # Создаем DataFrame для сравнения
    comparison_data = []
    
    for topic in ['кофе', 'чай']:
        df = data_frames[topic]
        comparison_data.append({
            'Тема': topic,
            'Посты': len(df),
            'Лайки': df['likes'].mean(),
            'Репосты': df['reposts'].mean(),
            'Комментарии': df['comments'].mean()
        })
    
    df_comparison = pd.DataFrame(comparison_data)
    
    # Интерактивный график
    fig = px.bar(df_comparison.melt(id_vars=['Тема'], 
                                     value_vars=['Лайки', 'Репосты', 'Комментарии']),
                 x='variable', y='value', color='Тема',
                 barmode='group',
                 title='Сравнение активности: Кофе vs Чай',
                 labels={'variable': 'Метрика', 'value': 'Среднее значение'})
    
    fig.show()
```
